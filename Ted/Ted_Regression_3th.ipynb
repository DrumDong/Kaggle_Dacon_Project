```
X= np.array(df['Inspiring'])
y= np.array(df['views'])
print(X.shape,y.shape)
```

```
def linear_regression(X,y,w,b):
  w = np.zeros(X.shape[0]) +w
  b = np.zeros(X.shape[0]) +b
  y_real = w*X +b
  #print(np.square(y_real - y).shape)
  mse = sum(np.square(y_real - y))/len(y)
  
  
  print('회귀계수는:',w.mean())
  print('회귀절편은:',b.mean())

  print('mse:',mse)
  

  return w,b,mse,y_real

#w,b,mse,y_real = linear_regression(X,y)

def numerical_gradient(X,y,y_real,w,b,eta=0.01):
  new_w = w - eta*(2/len(y))*np.dot(X.reshape(-1,2550),(y_real-y).reshape(2550,-1))
  # shape: (1,2550)
  new_b = b - eta*(2/len(y))*sum(y_real-y)                                          
  # shape: (2550,)
  
  return new_w,new_b

def gradient_descent(X,y,w,b,epoch):
  w_list = []
  b_list =[]
  mse_list =[]

  #초기 회귀계수(w),회귀절편(b) 설정 후 mse 출력
  w,b,mse,y_real = linear_regression(X,y,w,b)
  w_list.append(w.mean())
  b_list.append(b.mean())
  mse_list.append(mse)
  
  for epoch in range(1,epoch):
    print('epoch:',epoch)
    if mse > 0:
      new_w,new_b=numerical_gradient(X,y,y_real,w,b)
      
      #하나의 값만 받기위해 평균처리 min(),max()해도 다 같음.
      w= new_w.mean() 
      b= new_b.mean()

      #갱신된 w와 b를 회귀식에 대입
      w,b,mse,y_real=linear_regression(X,y,w,b)
      w_list.append(w.mean())
      b_list.append(b.mean())
      mse_list.append(mse)
    if mse == float('inf'):
      break
  return w_list,b_list,mse_list
```
