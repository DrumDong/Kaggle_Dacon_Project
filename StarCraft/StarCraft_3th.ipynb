```
def make_df(toy_data,id):
  st_axis_0=np.array(toy_data[(toy_data['game_id']==id)&
           (toy_data['player']==0)&
           (toy_data['event']=='Camera')][['x_axis','y_axis']])[0]
         
  st_axis_1=np.array(toy_data[(toy_data['game_id']==id)&
           (toy_data['player']==1)&
           (toy_data['event']=='Camera')][['x_axis','y_axis']])[0]
  
  # 계산
  df_abs_0=abs(toy_data[(toy_data['game_id']==id)&
           (toy_data['player']==0)&
           (toy_data['event']=='Camera')][['x_axis','y_axis']][-200:] - st_axis_0)

  df_abs_1=abs(toy_data[(toy_data['game_id']==id)&
           (toy_data['player']==1)&
           (toy_data['event']=='Camera')][['x_axis','y_axis']][-200:] - st_axis_1)
  
  df_abs_0['x+y']= df_abs_0['x_axis']+df_abs_0['y_axis']
  df_abs_1['x+y']= df_abs_1['x_axis']+df_abs_1['y_axis']

  from collections import OrderedDict
  data = OrderedDict()

  for idx,i in enumerate( np.array(df_abs_0['x+y'])):
    data['0_%s'%(idx)]=[i]
  for idx,i in enumerate( np.array(df_abs_1['x+y'])):
    data['1_%s'%(idx)]=[i]
  
  final_df = pd.DataFrame(data,columns=data.keys())
  
  #print(final_df)
  return final_df 
```

```
make_df(toy_data,100)
```

한 줄씩 나오는 데이터 프레임을 행삽입하여 만들자.

```
df= pd.DataFrame(columns=make_df(toy_data,100).columns)

for num in set(toy_data['game_id'].values):
  df = pd.concat([df, make_df(toy_data,num)], axis=0,sort=False)
```
